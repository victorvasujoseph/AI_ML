```python
import matplotlib.pyplot as plt

# Sample data - list of salaries
salaries = [30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000,
            80000, 85000, 90000, 95000, 100000, 110000, 120000, 130000, 140000, 150000]

# Plot histogram
plt.hist(salaries, bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Salary Range')
plt.ylabel('Frequency')
plt.title('Salary Distribution')

mean_salary = np.mean(salaries)
std_salary = np.std(salaries)

# Annotate mean and standard deviation
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=1)
plt.annotate(f'Mean: ${mean_salary:.2f}', xy=(mean_salary, 0), xytext=(mean_salary + 15000, 3),
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate(f'Standard Deviation: ${std_salary:.2f}', xy=(mean_salary, 0), xytext=(mean_salary + 15000, 6),
             arrowprops=dict(facecolor='blue', shrink=0.05))

# Show plot
plt.show()
```
