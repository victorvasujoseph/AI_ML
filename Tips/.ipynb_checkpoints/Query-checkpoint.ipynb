```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/world_population.csv', na_values =['N/A', 'Missing', 'Not available','?', 'NA'])
#In this example, 'N/A', 'Missing', and 'Not available' would be treated as NaN values when reading the CSV file into a DataFrame. 
#This helps ensure consistency in handling missing data across different datasets.
```

```python
np.random.seed(42)
#np.random.seed(42) sets the random seed for NumPy's random number generator. 
#This ensures reproducibility of random number generation.
# You can also set index column by doing 
# df = pd.read_csv('data/world_population.csv', index_col = "Rank")
```

```python
df = df.reindex(np.random.permutation(df.index))
```

```python
df.sample(10)
```

```python
df.reset_index(inplace=True,drop=True)
'''df: This is the DataFrame you're working with..reset_index(): This method resets the index of the DataFrame. By default, it moves the current index to a new column called 'index' and sets a new integer index.
inplace=True: This parameter specifies that the operation should be performed on the DataFrame itself, rather than returning a new DataFrame. If inplace is set to True, the method modifies the DataFrame in place and returns None.
drop=True: This parameter specifies whether to drop the old index. If set to True, the old index will be dropped. In this case, drop=True is specified, so the old index will be dropped.'''
```
