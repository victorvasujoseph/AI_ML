```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace this with your DataFrame)
data = {
    'player': ['anderga0', 'colonba0', 'davanje0', 'donnebr0', 'eckstda0', 'erstada0', 'escobke0', 'figgich0', 'glaustr0', 'greggke0'],
    'salary': [6200000, 11000000, 375000, 375000, 2150000, 7750000, 5750000, 320000, 9900000, 301500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract salary data
salaries = df['salary']

# Plot histogram
plt.hist(salaries, bins=5, color='skyblue', edgecolor='black', density=False)

# Add labels and title
plt.xlabel('Salary')
plt.ylabel('Number of Players')
plt.title('Salary Distribution')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace this with your DataFrame)
data = {
    'player': ['anderga0', 'colonba0', 'davanje0', 'donnebr0', 'eckstda0', 'erstada0', 'escobke0', 'figgich0', 'glaustr0', 'greggke0'],
    'salary': [6200000, 11000000, 375000, 375000, 2150000, 7750000, 5750000, 320000, 9900000, 301500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract salary data
salaries = df['salary']

# Define bins for salary ranges
salary_bins = [0, 2000000, 4000000, 6000000, 8000000, 10000000, 12000000]

# Plot histogram with specified bins
plt.hist(salaries, bins=salary_bins, color='skyblue', edgecolor='black', density=False)

# Add labels and title
plt.xlabel('Salary Range')
plt.ylabel('Number of Players')
plt.title('Salary Distribution')

# Show plot
plt.show()
```
