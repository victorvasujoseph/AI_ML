```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names() #inbuild data set
```

```python
dfCarCrash = sns.load_dataset('car_crashes')
```

```python
dfCarCrash.head(10)
```

```python
# press tab after . to see the list of functions and shift + tab on a function to see the signature
```

```python
sns.histplot(dfCarCrash['not_distracted'], bins=20)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=dfCarCrash, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=dfCarCrash, kind='hex')
```

```python
sns.kdeplot(dfCarCrash['speeding'])
```

```python
dfTips = sns.load_dataset('tips')
dfTips
```

```python
sns.pairplot(dfTips, hue='sex')
```

```python
sns.pairplot(dfTips, hue='smoker')
```

```python
sns.set_style('ticks') # ticks, darkgrid, whitegrid
sns.set_context('paper', font_scale=1.4) # try = 'talk' or 'poster'
plt.figure(figsize=(10,4))
sns.jointplot(x='speeding', y='alcohol', data=dfCarCrash, kind='reg')
#sns.despine(left=True,bottom=True) # Defines the x & y axis line
```

```python
sns.barplot(x='sex', y='total_bill', data=dfTips, estimator=np.mean)
```

```python
sns.boxplot(x='day', y='total_bill', data=dfTips, hue='sex')
plt.legend(loc=0)
```

```python
sns.violinplot(x='day', y='total_bill', data=dfTips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=dfTips, hue='sex', jitter=True, dodge =True) # jitter=True spaces the plots.
# try without hue
```

```python
sns.stripplot(x='day', y='total_bill', data=dfTips, hue='sex', jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=dfTips)
```

```python
plt.figure(figsize=(10,5))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=dfTips, hue='sex', palette='magma') 
# check out palette in https://matplotlib.org/cheatsheets/ and use it for various color maps
plt.legend(loc=2)
```

```python
sns.set_style('white')
sns.set_context('paper')
sns.heatmap(dfCarCrash.corr(numeric_only=True),annot=True,cmap='Blues') #numeric_only=True will skip 
```

```python
flights = sns.load_dataset('flights')
xflight = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(xflight, cmap='Blues', linecolor='white', linewidth='1')
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species')
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(plt.scatter)
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species',x_vars = ["sepal_length","sepal_width"], y_vars = ["petal_length","petal_width"])
#iris_g.map_diag(plt.hist)
iris_g.map(plt.scatter)
iris_g.add_legend() # Adds legend
```

```python
tipsfg = sns.FacetGrid(dfTips,col='time', hue='smoker', height=4, aspect=1.3, col_order=['Dinner','Lunch'], palette='magma')
tipsfg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```
