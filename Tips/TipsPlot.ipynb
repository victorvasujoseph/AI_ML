```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate population data (non-normal distribution)
#population = pd.Series(np.random.exponential(scale=50000, size=10000))  # Exponential distribution for example
baseball = pd.read_csv('/Users/victorjoseph/UCB_BK/MLALUCB/Tips/data/baseball.csv')
population = baseball['salary']

# Step 2 and 3: Take random samples of various sizes and calculate the mean
sample_sizes = [10,20,30,50,100,200]  # Different sample sizes
sample_means = []

for size in sample_sizes:
    sample_means.append([population.sample(size).mean() for _ in range(10)])  # Take 1000 samples for each size

# Step 4: Plot the distribution of sample means
plt.figure(figsize=(12, 8))

for i, size in enumerate(sample_sizes):
    plt.subplot(2, 2, i + 1)
    plt.hist(sample_means[i], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Sample Size = {size}')
    plt.xlabel('Mean Salary')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate population data (non-normal distribution)
#population = pd.Series(np.random.exponential(scale=50000, size=10000))  # Exponential distribution for example
baseball = pd.read_csv('/Users/victorjoseph/UCB_BK/MLALUCB/Tips/data/baseball.csv')
population = baseball['salary']

# Step 2 and 3: Take random samples of various sizes and calculate the mean
sample_sizes = [50, 100, 300, 700]  # Different sample sizes
sample_means = []

for size in sample_sizes:
    sample_means.append([population.sample(size).mean() for _ in range(1000)])  # Take 1000 samples for each size

# Step 4: Plot the distribution of sample means
plt.figure(figsize=(12, 8))

for i, size in enumerate(sample_sizes):
    plt.subplot(2, 2, i + 1)
    plt.hist(sample_means[i], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Sample Size = {size}')
    plt.xlabel('Mean Salary')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
